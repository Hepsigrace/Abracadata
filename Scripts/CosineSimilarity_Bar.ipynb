```python
# Read file
import os
import pandas as pd
from google.colab import files

uploaded = files.upload()

# Assuming the uploaded file is named
uploaded_file_path = list(uploaded.keys())[0]

# Read CSV into DataFrame
try:
    dfimported = pd.read_csv(uploaded_file_path, encoding='utf-8')
except UnicodeDecodeError:
    dfimported = pd.read_csv(uploaded_file_path, encoding='ISO-8859-1')

# Display the DataFrame
print(dfimported)
```

```python
import matplotlib.pyplot as plt

# Cosine Similarity Accuracy
csgem = dfimported['LabelGemini'].sum()/len(dfimported)
csgpt = dfimported['LabelGPT'].sum()/len(dfimported)


cs_scores = {'Gemini Pro': csgem, 'GPT 3.5': csgpt}

plt.figure(figsize=(6, 4))
bars = plt.bar(cs_scores.keys(), cs_scores.values(), color=['blue', 'green'])
plt.title('Cosine Similarity Accuracy')
plt.xlabel('LLMs')
plt.ylabel('Accuracy')

# Annotate bars with accuracy
for llm, score in cs_scores.items():
    plt.text(llm, score, str(round(score*100,2))+'%', ha='center', va='bottom')

plt.show()
```
